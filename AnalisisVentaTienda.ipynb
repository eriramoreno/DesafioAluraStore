### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
#calcular el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.

ingreso_total_tienda1 = tienda['Precio'].sum()
ingreso_total_tienda2 = tienda2['Precio'].sum()
ingreso_total_tienda3 = tienda3['Precio'].sum()
ingreso_total_tienda4 = tienda4['Precio'].sum()

print(f"Ingreso total Tienda 1: {ingreso_total_tienda1}")
print(f"Ingreso total Tienda 2: {ingreso_total_tienda2}")
print(f"Ingreso total Tienda 3: {ingreso_total_tienda3}")
print(f"Ingreso total Tienda 4: {ingreso_total_tienda4}")
```

```python
#grafico para complementar este analisis

import matplotlib.pyplot as plt

tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
ingresos = [ingreso_total_tienda1, ingreso_total_tienda2, ingreso_total_tienda3, ingreso_total_tienda4]

plt.figure(figsize=(8, 6))
plt.bar(tiendas, ingresos, color=['blue', 'green', 'red', 'purple'])
plt.xlabel('Tienda')
plt.ylabel('Ingreso Total')
plt.title('Ingreso Total por Tienda')
plt.grid(axis='y')
plt.show()
```

# 2. Ventas por categoría

```python
#cantidad de productos vendidos por categoria en cada tienda

print("\nFrecuencia de Categoría del Producto por Tienda:")
print("Tienda 1:")
print(tienda['Categoría del Producto'].value_counts())
print("\nTienda 2:")
print(tienda2['Categoría del Producto'].value_counts())
print("\nTienda 3:")
print(tienda3['Categoría del Producto'].value_counts())
print("\nTienda 4:")
print(tienda4['Categoría del Producto'].value_counts())
```

# 3. Calificación promedio de la tienda


```python
#Calificación promedio de cada una de las tiendas

calificacion_promedio_tienda1 = tienda['Calificación'].mean()
calificacion_promedio_tienda2 = tienda2['Calificación'].mean()
calificacion_promedio_tienda3 = tienda3['Calificación'].mean()
calificacion_promedio_tienda4 = tienda4['Calificación'].mean()

print(f"\nCalificación promedio Tienda 1: {calificacion_promedio_tienda1}")
print(f"Calificación promedio Tienda 2: {calificacion_promedio_tienda2}")
print(f"Calificación promedio Tienda 3: {calificacion_promedio_tienda3}")
print(f"Calificación promedio Tienda 4: {calificacion_promedio_tienda4}")
```

```python
#grafico de Calificacion promedio por tienda.

import matplotlib.pyplot as plt
calificaciones = [calificacion_promedio_tienda1, calificacion_promedio_tienda2, calificacion_promedio_tienda3, calificacion_promedio_tienda4]

plt.figure(figsize=(8, 6))
plt.plot(tiendas, calificaciones, marker='o', linestyle='-', color='orange')
plt.xlabel('Tienda')
plt.ylabel('Calificación Promedio')
plt.title('Calificación Promedio por Tienda')
plt.ylim(0, 5) # Assuming a rating scale of 0-5
plt.grid(True)
plt.show()
```

# 4. Productos más y menos vendidos

```python
#Producto más y menos vendido en cada tienda.

# Agrupar por 'Producto' y contar las ventas para cada tienda
ventas_producto_tienda1 = tienda['Producto'].value_counts()
ventas_producto_tienda2 = tienda2['Producto'].value_counts()
ventas_producto_tienda3 = tienda3['Producto'].value_counts()
ventas_producto_tienda4 = tienda4['Producto'].value_counts()

# Producto más vendido en cada tienda
producto_mas_vendido_tienda1 = ventas_producto_tienda1.idxmax()
producto_mas_vendido_tienda2 = ventas_producto_tienda2.idxmax()
producto_mas_vendido_tienda3 = ventas_producto_tienda3.idxmax()
producto_mas_vendido_tienda4 = ventas_producto_tienda4.idxmax()

# Producto menos vendido en cada tienda
producto_menos_vendido_tienda1 = ventas_producto_tienda1.idxmin()
producto_menos_vendido_tienda2 = ventas_producto_tienda2.idxmin()
producto_menos_vendido_tienda3 = ventas_producto_tienda3.idxmin()
producto_menos_vendido_tienda4 = ventas_producto_tienda4.idxmin()

print("\nProductos más y menos vendidos por Tienda:")
print(f"Tienda 1 - Más vendido: {producto_mas_vendido_tienda1}, Menos vendido: {producto_menos_vendido_tienda1}")
print(f"Tienda 2 - Más vendido: {producto_mas_vendido_tienda2}, Menos vendido: {producto_menos_vendido_tienda2}")
print(f"Tienda 3 - Más vendido: {producto_mas_vendido_tienda3}, Menos vendido: {producto_menos_vendido_tienda3}")
print(f"Tienda 4 - Más vendido: {producto_mas_vendido_tienda4}, Menos vendido: {producto_menos_vendido_tienda4}")
```

# 5. Envío promedio por tienda

```python
#Costo de envío promedio de cada tienda

costo_envio_promedio_tienda1 = tienda['Costo de envío'].mean()
costo_envio_promedio_tienda2 = tienda2['Costo de envío'].mean()
costo_envio_promedio_tienda3 = tienda3['Costo de envío'].mean()
costo_envio_promedio_tienda4 = tienda4['Costo de envío'].mean()

print("\nCosto de Envío promedio por Tienda:")
print(f"Tienda 1: {costo_envio_promedio_tienda1}")
print(f"Tienda 2: {costo_envio_promedio_tienda2}")
print(f"Tienda 3: {costo_envio_promedio_tienda3}")
print(f"Tienda 4: {costo_envio_promedio_tienda4}")
```
